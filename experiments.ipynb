```python
categories = parallel_sorting["1"].keys()
legend = {}
for cat in categories:
  legend[cat] = [(int(key), parallel_sorting["1"][cat] / elem[cat]) for key, elem in parallel_sorting.items()]

fig = plt.figure()
ax = plt.subplot(111)
for key, values in legend.items():
  if key == 'iters': continue
  values_x, values_y = zip(*values)
  ax.plot(values_x, values_y, label=key)
box = ax.get_position()
fig.set_size_inches(18.5, 10.5)
ax.set_position([box.x0, box.y0, box.width * 0.6, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.suptitle('Parallelism Plot for Large Data', fontsize=20)
plt.xticks(np.arange(0, 17, 1.0))
plt.yticks(np.arange(0, 20, 1.0))
plt.xlabel('Cores', fontsize=18)
plt.ylabel('Parallelism', fontsize=16)
```
